```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import sys
sys.path.append('..')
from src.data.load_data import load_resumes, load_jobs
from src.data.preprocess import clean_text
```

```python
resumes = load_resumes()
jobs = load_jobs()
print("Resumes shape:", resumes.shape)
print("Jobs shape:", jobs.shape)
```

```python
plt.figure(figsize=(12,6))
resumes['Category'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Resume Categories Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/figures/resume_categories.png')
plt.show()
```

```python
plt.figure(figsize=(12,6))
jobs['title'].value_counts().head(20).plot(kind='bar', color='coral')
plt.title('Top 20 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/figures/top_jobs.png')
plt.show()
```

```python
all_resume_text = " ".join(resumes['Resume'].dropna().apply(clean_text))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_resume_text)
plt.figure(figsize=(14,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Resumes')
plt.tight_layout()
plt.savefig('../reports/figures/resume_wordcloud.png')
plt.show()
```

```python
print("=== RESUME DATASET ===")
print(resumes.describe())
print("\nCategories:\n", resumes['Category'].value_counts())

print("\n=== JOBS DATASET ===")
print(jobs.describe())
print("\nNull values:\n", jobs.isnull().sum())
```
